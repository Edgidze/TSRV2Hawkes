```python
def DGP1(seed, n):
    mu, sigma2 = gen_mu_sigma2(t, 'flat', seed)
    
    BR = 0
    points = gen_Cox(t, mu, seed, n)
    
    return points, mu, sigma2, BR


def DGP2(seed, n):
    mu, sigma2 = gen_mu_sigma2(t, 'U-shape', seed)
    
    BR = 0
    points = gen_Cox(t, mu, seed, n)
    
    return points, mu, sigma2, BR


def DGP3(seed, n):
    mu, sigma2 = gen_mu_sigma2(t, 'CIR', seed)
    mu += get_burst(mu, seed)
    
    BR = 0
    points = gen_Cox(t, mu, seed, n)
    
    return points, mu, sigma2, BR


def DGP4(seed, n):
    mu, sigma2 = gen_mu_sigma2(t, 'flat', seed)
    
    BR  = 0.8
    mu *= 1 - BR
    sigma2 *= (1 - BR)**2
    beta   = 2.0
    alpha  = BR * beta
    points = gen_HawkesExpKernels(
        t, mu, alpha=alpha, beta=beta, 
        n=n, seed=seed)
    
    return points, mu, sigma2, BR


def DGP5(seed, n):
    mu, sigma2 = gen_mu_sigma2(t, 'U-shape', seed)
    
    BR  = 0.8
    mu *= 1 - BR
    sigma2 *= (1 - BR)**2
    beta   = 2.0
    alpha  = BR * beta
    points = gen_HawkesExpKernels(
        t, mu, alpha=alpha, beta=beta, 
        n=n, seed=seed)
    
    return points, mu, sigma2, BR


def DGP6(seed, n):
    mu, sigma2 = gen_mu_sigma2(t, 'CIR', seed)
    mu += get_burst(mu, seed)
    
    BR  = 0.8
    mu *= 1 - BR
    sigma2 *= (1 - BR)**2
    beta   = 2.0
    alpha  = BR * beta
    points = gen_HawkesExpKernels(
        t, mu, alpha=alpha, beta=beta, 
        n=n, seed=seed)
    
    return points, mu, sigma2, BR

def DGP7(seed, n):
    mu, sigma2 = gen_mu_sigma2(t, 'flat', seed)
    
    BR  = 0.8
    mu *= 1 - BR
    sigma2  *= (1 - BR)**2
    cutoff   = 1
    exponent = 1 + 2
    multiplier = BR / cutoff**(1-exponent) * (exponent - 1)
    points = gen_HawkesPowerKernels(
        t, mu, multiplier=multiplier, 
        exponent=exponent, cutoff=cutoff,
        n=n, seed=seed)
    
    return points, mu, sigma2, BR


def DGP8(seed, n):
    mu, sigma2 = gen_mu_sigma2(t, 'U-shape', seed)
    
    BR  = 0.8
    mu *= 1 - BR
    sigma2  *= (1 - BR)**2
    cutoff   = 1
    exponent = 1 + 2
    multiplier = BR / cutoff**(1-exponent) * (exponent - 1)
    points = gen_HawkesPowerKernels(
        t, mu, multiplier=multiplier, 
        expoent=exponent, cutoff=cutoff,
        n=n, seed=seed)
    
    return points, mu, sigma2, BR


def DGP9(seed, n):
    mu, sigma2 = gen_mu_sigma2(t, 'CIR', seed)
    #mu += get_burst(mu, seed)  # For power-kernel, burst takes too long time.
    
    BR  = 0.8
    mu *= 1 - BR
    sigma2  *= (1 - BR)**2
    cutoff   = 1
    exponent = 1 + 2
    multiplier = BR / cutoff**(1-exponent) * (exponent - 1)
    points = gen_HawkesPowerKernels(
        t, mu, multiplier=multiplier, 
        exponent=exponent, cutoff=cutoff,
        n=n, seed=seed)
    
    return points, mu, sigma2, BR


def DGP10(seed, n):
    mu, sigma2 = gen_mu_sigma2(t, 'flat', seed)
    mu += get_burst(mu, seed)
    
    n_tmp = np.sqrt(n)  # For near-critical Hawkes, we use sqrt(n).
    BR = 1 - n_tmp**(-1) 
    points = np.array([])
    it = 0
    while np.abs(len(points) - n)/n > 0.1:
        beta   = 2.0
        alpha  = BR * beta
        points = gen_HawkesExpKernels(
            t, mu, alpha=alpha, beta=beta, 
            n=n_tmp, seed=seed + it, 
            force_simulation=True)
        it += 999999
    
    return points, mu, sigma2, BR


def gen_points(seed):
    t1 = time()
    np.random.seed(seed)

    out = []
    for n in ns:
        
        # Generation
        points, mu, sigma2, BR = DGP(seed, n)

        Mean    = n / (1 - BR) * np.sum(mu) * dt
        IV      = 1 / (1 - BR)**2 * np.sum(sigma2) * dt    
        
        # Save     
        out.append({'points': points, 'BR': BR, 'Mean': Mean, 'IV': IV})
            
        del points
        gc.collect()
        
    msg = "  %4d: %.3f sec"%(seed, time() - t1)
    info(msg)
    if VERBOSE: print(msg)    

    return out
```

```python
######### Parameters ##########
N_sim   = 1000
ns      = [50000, 150000, 1000000]
VERBOSE = False
###############################

dt = 1 / (np.max(ns) * 5)
t  = np.arange(0, 1.0, dt)
```

## Generate points

```python
%%time
def log_out(i, out):
    pool_out[i] = out
    
# WARNING: DGP7-DGP9 takes 40~80 hours per model.
DGPs = [DGP1, DGP2, DGP3, DGP4, DGP5, DGP6, DGP7, DGP8, DGP9, DGP10]

for DGP in DGPs:
    print(DGP.__name__); info(DGP.__name__)
    
    pool_out = {}
    pool = Pool(cpu_count()-8) 
    try:
        for seed in range(N_sim):
            if seed not in pool_out:
                pool.apply_async(gen_points, args=(seed,), 
                                 callback=partial(log_out, seed))
        pool.close()
        pool.join()
    except KeyboardInterrupt:
        pool.terminate()
    pool_out = dict(sorted(pool_out.items()))
    
    # Save
    filename  = 'save/points/%s.dat'%DGP.__name__
    header    = {"N_sim": N_sim, "ns": ns, "t": t, "DGP":DGP}
    joblib.dump({'header': header, 'data': pool_out}, filename)

alarm()
```

## Estimation

```python
def estimate(seed):
    t1 = time()
    np.random.seed(seed)

    out = []
    for n in ns:
        
        # Load Data
        try:
            ni  = ns.index(n)
            dat = save_points['data'][seed][ni]
        except:
            lst = ['BR', 'Mean', 'IV']
            dat = dict([('points',np.array([]))] + [(i,np.nan) for i in lst])
            
        points  = dat['points'] 
        BR      = dat['BR']
        Mean    = dat['Mean'] 
        IV      = dat['IV']
        if DGP.__name__ == 'DGP9':
            mu, _ = gen_mu_sigma2(t, 'CIR', seed)
            mu_burst = get_burst(mu, seed)
            if isinstance(mu_burst, int):
                burst_points = []
            else:
                burst_points = gen_Cox(t, mu_burst, seed, n=n)
            points = np.sort(np.append(points, burst_points))
            Mean += n * np.sum(mu_burst) * dt
        
        # Estimation
        BR_tmp = BR
        if DGP.__name__ == 'DGP10':
            res = BR_estimation(points, truncation=False)
            # For near-critical case, we cannot distinguish BR=0.999 from BR=0.
            BR_tmp  = 0  
        else:
            res = BR_estimation(points, truncation=True)
        
        BR_test    = res['BR'] / res['SD(BR)']
        SD_int_int = res['SD(Mean)'] if BR_test > 1.96 else np.sqrt(res['N'])
        z_int_int  = (res['Mean'] - Mean) / SD_int_int
        
        AVar_Mean  = Mean / (1 - BR)**2 / n
        EstAVar_Mean_Cox = res['N'] / n
        EstAVar_Mean_Haw = res['SD(Mean)']**2 / n
        
        out.append(
            {'BR'         : BR,
             'Mean'       : Mean,
             'IV'         : IV,
             'z_Mean'     : (res['Mean'] - Mean)   / res['SD(Mean)'],
             'z_BR'       : (res['BR'] - BR_tmp)   / res['SD(BR)'],
             'z_BR_const' : (res['BR_const'] - BR) / res['SD(BR_const)'],
             'z_int_int'  : z_int_int,
             'AVar_Mean'  : AVar_Mean,
             'EstAVar_Mean_Cox': EstAVar_Mean_Cox,
             'EstAVar_Mean_Haw': EstAVar_Mean_Haw,
             'z_IV'       : (res['IV'] * res['N']**2 - IV * n**2) \
                             / (res['SD(IV)'] * res['N']**2),
             'res'        : res,
            })
            
        del points
        gc.collect()
        
    msg = "  %4d: %.3f sec"%(seed, time() - t1)
    info(msg)
    if VERBOSE: print(msg)
    
    return out
```

```python
%%time
def log_out(i, out):
    pool_out[i] = out

DGPs = [DGP1, DGP2, DGP3, DGP4, DGP5, DGP6, DGP7, DGP8, DGP9, DGP10]

pool_outs = {}
for DGP in DGPs:
    print(DGP.__name__); info(DGP.__name__)
    filename = 'save/points/%s.dat'%DGP.__name__
    try:
        save_points = joblib.load(filename)
    except:
        save_points = None
    
    pool_out = {}
    pool = Pool(cpu_count()-10) 
    try:
        for seed in range(N_sim):
            if seed not in pool_out:
                pool.apply_async(estimate, args=(seed,), 
                                 callback=partial(log_out, seed))
        pool.close()
        pool.join()
    except KeyboardInterrupt:
        pool.terminate()
    pool_out = dict(sorted(pool_out.items()))
    
    pool_outs[DGP.__name__] = pool_out

alarm()
```

## Comparison

```python
def comparison(seed):
    t1 = time()
    np.random.seed(seed)

    out = []
    for n in ns:
        
        # Load Data
        try:
            ni  = ns.index(n)
            dat = save_points['data'][seed][ni]
        except:
            lst = ['BR', 'Mean', 'IV']
            dat = dict([('points',np.array([]))] + [(i,np.nan) for i in lst])
            
        points  = dat['points'] 
        BR      = dat['BR']
        Mean    = dat['Mean'] 
        IV      = dat['IV']
        if DGP.__name__ == 'DGP9':
            mu, _ = gen_mu_sigma2(t, 'CIR', seed)
            mu_burst = get_burst(mu, seed)
            burst_points = gen_Cox(t, mu_burst, seed, n=n)
            points = np.sort(np.append(points, burst_points))
            Mean += n * np.sum(mu_burst) * dt
        
        # Estimation
        ress = {}
        for estimator in estimators:
            ress[estimator.__name__] = estimator(points)
        
        out.append({'BR': BR, 'ress': ress})
            
        del points
        gc.collect()
        
    msg = "  %4d: %.3f sec"%(seed, time() - t1)
    info(msg)
    if VERBOSE: print(msg)
    
    return out
```

```python
estimators = [
    BR_estimation,
    Hardiman,
    AVG_Hardiman_4,
    Exp_estimation,
    EM_estimation,
]
names = ['Our method', 'H&B', 'H&B(2h)', 'MLE(exp)', 'EM']

######### Parameters ##########
N_sim   = 1000
ns      = [50000, 150000]
VERBOSE = True
###############################

dt = 1 / (np.max(ns) * 5)
t  = np.arange(0, 1.0, dt)
```

```python
%%time
def log_out(i, out):
    pool_out[i] = out

DGPs = [DGP7, DGP8, DGP9] #[DGP1, DGP2, DGP3, DGP4, DGP5, DGP6, DGP7, DGP8, DGP9, DGP10]  

#pool_outs = {}
for DGP in DGPs:
    print(DGP.__name__); info(DGP.__name__)
    filename = 'save/points/%s.dat'%DGP.__name__
    try:
        save_points = joblib.load(filename)
    except:
        save_points = None
    
    pool_out = {}
    pool = Pool(cpu_count()-8) 
    try:
        for seed in range(N_sim):
            if seed not in pool_out:
                pool.apply_async(comparison, args=(seed,), 
                                 callback=partial(log_out, seed))
        pool.close()
        pool.join()
    except KeyboardInterrupt:
        pool.terminate()
    pool_out = dict(sorted(pool_out.items()))
    
    pool_outs[DGP.__name__] = pool_out

alarm()
```
